```python
# Step 2: Create a vocabulary
vocab = sorted(set(tokens))
stoi = {word: i for i, word in enumerate(vocab)}
itos = {i: word for word, i in stoi.items()}
vocab_size = len(vocab)
```

```python
# Step 3: Generate training samples
def train_sample(block_size):
  X, Y = [], []
  for i in range(len(tokens) - block_size):
      context = [stoi[tokens[i+j]] for j in range(block_size)]
      target = stoi[tokens[i + block_size]]
      X.append(context)
      Y.append(target)
      # training_data.append((context, target))
  X = torch.tensor(X).to(device)
  Y = torch.tensor(Y).to(device)
  return X,Y
```

```python
# Embedding layer for the context
emb_dim = 150
emb = torch.nn.Embedding(len(stoi), emb_dim)
```

### Plotting the embedding for top 100 most meaningful words in the dataset

```python

tagged_words = pos_tag(vocab)
adjectives_and_nouns = [word for word, pos in tagged_words if pos.startswith('JJ') or pos.startswith('NN')]

adjectives_and_nouns
```

```python
word_freq = Counter(adjectives_and_nouns)
common_words = word_freq.most_common(150)
common_words = [word for word, _ in common_words]
stop_words = set(stopwords.words('english'))
common_words = [word for word in common_words if word not in stop_words]
common_words = common_words[5:]
```

```python
word_vector = emb.weight.detach().numpy()
vectors_of_interest = np.array([word_vector[stoi[word]] for word in common_words])
```

**Visualising our vocabulary using T-sne**

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=3, n_iter=1000)
tsne_embeddings = tsne.fit_transform(vectors_of_interest)

# Plot t-SNE embeddings with different colors
plt.figure(figsize=(15, 15))
for i, word in enumerate(common_words):
    plt.scatter(tsne_embeddings[i, 0], tsne_embeddings[i, 1], marker='.')  # Assign different color for each point
    plt.annotate(word, (tsne_embeddings[i, 0], tsne_embeddings[i, 1]))  # Annotate each point with the word
plt.title('t-SNE Embeddings of Word Embeddings from Shakespeare\'s Plays')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
# plt.legend()  # Show legend with labels
plt.show()
```

```python
class NextChar(nn.Module):
  def __init__(self, block_size, vocab_size, emb_dim, hidden_size):
    super().__init__()
    self.emb = nn.Embedding(vocab_size, emb_dim)
    self.lin1 = nn.Linear(block_size * emb_dim, hidden_size)
    self.lin2 = nn.Linear(hidden_size, hidden_size)
    self.lin3 = nn.Linear(hidden_size, vocab_size)

  def forward(self, x):
    x = self.emb(x)
    x = x.view(x.shape[0], -1)
    x = torch.sin(self.lin1(x))
    x = torch.sin(self.lin2(x))
    x = self.lin3(x)
    return x
```

```python
def generate_text(model, itos, stoi, block_size, context ,max_len=10):
    name = ''
    for i in range(max_len):
        x = torch.tensor(context).view(1, -1).to(device)
        y_pred = model(x)
        ix = torch.distributions.categorical.Categorical(logits=y_pred).sample().item()
        ch = itos[ix]
        name =name+' '+ch
        context = context[1:] + [ix]
    return name

```

```python
# Train the model
def train(model,X,Y):
    loss_fn = nn.CrossEntropyLoss()
    opt = torch.optim.AdamW(model.parameters(), lr=0.001)
    import time
    # Mini-batch training
    batch_size = 4096
    print_every = 100
    elapsed_time = []
    for epoch in range(500):
        for i in range(0, X.shape[0], batch_size):
            x = X[i:i+batch_size]
            y = Y[i:i+batch_size]
            y_pred = model(x)
            loss = loss_fn(y_pred, y)
            loss.backward()
            opt.step()
            opt.zero_grad()
        if epoch % print_every == 0:
            print(epoch, loss.item())
```
